```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
R = L = 1
I_0 = 1

# corrente em funcao do tempo
t = np.linspace(-1,5,501)
i_neg = np.vectorize(lambda x: I_0)(t[t < 0])
i_pos = np.vectorize(lambda x: I_0 * np.exp(-R*x/L))(t[t >= 0])
i = np.concatenate((i_neg, i_pos))
plt.plot(t, i, 'k')
plt.show()

# tensão em função do tempo
v_neg = np.vectorize(lambda x: 0)(t[t < 0])
v_pos = np.vectorize(lambda x: I_0 * np.exp(-R*x/L))(t[t >= 0])
v = np.concatenate((v_neg, v_pos))
plt.plot(t, v, 'k')
plt.show()

# potencia em função do tempo
p_neg = np.vectorize(lambda x: 0)(t[t < 0])
p_pos = np.vectorize(lambda x: I_0 * I_0 * np.exp(-2*R*x/L))(t[t >= 0])
p = np.concatenate((p_neg, p_pos))
plt.plot(t, p, 'k')
plt.show()

# energia em função do tempo
w_neg = np.vectorize(lambda x: 0)(t[t < 0])
w_pos = np.vectorize(lambda x: (1/2) * L * I_0 * I_0 * (1 - np.exp(-2*R*x/L)))(t[t >= 0])
w = np.concatenate((w_neg, w_pos))
plt.plot(t, w, 'k')
plt.show()
```
