**Letra a)**

No instante de tempo \\(t = 0\\), a chave é comutada da posição \\(a\\) para a posição \\(b\\). Até então, após um longo tempo, o indutor funcionava como um elemento de curto-circuito conectado à fonte de corrente de \\(8 A\\), portanto, apresenta \\(|i(0)| = 8A\\). Entretanto, como a corrente \\(i(0)\\) está num sentido oposto à corrente oriunda da fonte de tensão, logo: \\(i(0) = -8 A\\).

Sabemos que \\(V_S = 24V\\), \\(R = 2 \Omega\\) e \\(L = 0.2 \times 10^{-3} H\\)

Logo, aplicando estas variáveis à função da corrente

\\[
\begin{align} 
i(t) &= ({i(0) - {V_{S} \over R}}) \cdot e^{- {R \over L} t} + {V_{S} \over R} 
\end{align}
\\]

Temos que:

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import display, Markdown, Latex

i0, V_s, R, L = -8, 24, 2, 0.2

i_t = '\\\\(i(t) = {Vsr:.0f} {i0vsr:.0f} \cdot e^{{-{RL:.0f}t}}\\\\)'.format(
    i0vsr = i0 - V_s / R,
    Vsr = V_s / R,
    RL = R / L
)

display(Markdown(i_t))
```

Sim, pois, depois de longo tempo no estado \\(b\\) o indutor será um curto-circuito e sua tensão será nula já que sua corrente não mais variará.

Ainda, por possuir energia armazenada e receber da fonte \\(V_{S}\\) uma corrente que se sobrepõe ao sentido da corrente inicial \\(i(0)\\) que possui, o indutor faz uso de sua energia magnética armazenada para se opor á variação da corrente a qual é submetido.

Desta forma, \\(V_{L}\\) é dada pela contribuição da tensão da fonte de tensão sobreposta à contribuição inicial originária da fonte de corrente multiplicada por e^{-Rt \over L} que tende a zero quando \\(t \geq {{5L} \over R}\\).

**Letra e)**

```python
t = np.linspace(-0.25, 0.5, 1001)
i = np.vectorize(lambda x: -8 if x < 0 else (i0 - V_s/R)*np.exp(-R*x/L) + V_s/R)(t)
v = np.vectorize(lambda x: 0 if x < 0 else (V_s - R*i0)*np.exp(-R*x/L))(t)
# plotar corrente
plt.plot(t, i, label = 'Corrente')
# plotar tensao
plt.plot(t, v, label = 'Tensão')
# mostar as duas curvas
plt.legend()
plt.show()
```
